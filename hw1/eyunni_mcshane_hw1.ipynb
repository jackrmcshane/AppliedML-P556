Functions to find the mean and median of non-zero instances of the price attribute.

```python
class Stats():
    def __init__(self):
        self.v = []
        self.nzv()
        
    def nzv(self):
        for i in df['price']:
            if i != 0:
                self.v.append(i)
            else:
                continue
        return self.v

    def mean_price(self):
        return sum(self.v)/len(self.v)
    
    def median_price(self):
        sort_list = sorted(self.v)
        length_list = len(self.v)
        mid_index = (length_list)//2

        if length_list % 2:
            return sort_list[mid_index]
        else:
            return ( sort_list[mid_index] + sort_list[mid_index + 1] ) / 2
```

# H. Correlation Coeff

```python
import math
def squared_sum(col, count):
  return sum( [col[i]**2 for i in range(count)] )

def corr(feat, y):
  count = len(y)
  y_sum = sum(y)
  feat_sum = sum(feat)
  prodsum = sum( [feat[i] * y[i] for i in range(count)] )
  return ( (count * prodsum) - (y_sum * feat_sum) ) / math.sqrt( (count * squared_sum(feat, count) - feat_sum**2) * (count * squared_sum(y, count) - y_sum**2 ) )
```

# L. Evaluating Error

```python
## error functions definitions
def mse(true_vals: np.ndarray, predictions: np.ndarray):
  return sum( (true_vals - predictions) ** 2 ) / len(true_vals)

def mae(true_vals: np.ndarray, predictions: np.ndarray):
  return sum( abs( true_vals - predictions) ) / len(true_vals)
```

```python
# df error
df_mae_errors = list()
df_mse_errors = list()
for i in range(len(df_predictions)):
  df_mae_errors.append( mae(df_correct_vals[i].to_numpy(), df_predictions[i]) )
  df_mse_errors.append( mse(df_correct_vals[i].to_numpy(), df_predictions[i]) )

print('df Mean Absolute Error Values by Fold: ', *df_mae_errors)
print('df Mean Squared Error Values by Fold: ', *df_mse_errors)
```
